```python
import os
import random

def keep_only_n_images(folder_path, n=300):
    # Get a list of all files in the folder
    all_files = os.listdir(folder_path)

    # Filter only image files (you can customize the condition based on your file extensions)
    image_files = [file for file in all_files if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif'))]

    # Calculate the number of files to delete
    files_to_delete = len(image_files) - n

    if files_to_delete <= 0:
        print(f'No need to delete files. Current count: {len(image_files)}')
        return

    # Randomly select files to delete
    files_to_delete_list = random.sample(image_files, files_to_delete)

    # Delete the selected files
    for file_to_delete in files_to_delete_list:
        file_path = os.path.join(folder_path, file_to_delete)
        os.remove(file_path)
        print(f'Deleted: {file_path}')

    print(f'Kept {n} files. Deleted {files_to_delete} files.')

# Example usage
folder_path = 'D:\car_plate_detection\Arabic Letters & Numbers OCR\Dataset3\لا'
keep_only_n_images(folder_path, n=300)
```
